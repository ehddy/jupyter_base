```python
iphone_df.drop('iPhone XR', axis=0, inplace=False) 
```

**inplace=False**는 기존의 데이터프레임을 건들지 않고 새로운 dataframe을 생성  
데이터 프레임 자체를 바꾸고 싶으면 **inplace=True**

axis=1 = axis='columns'  
axis=0 = axis='index'
